# Hand Checking

```python
import random as r
from itertools import groupby,  combinations

suits = ["hearts", "clubs", "diamonds", "spades", "stars", "moons"]
names = [None,None,2,3,4,5,6,7,8,9,10,'jack','queen','king','ace']

class Card():

    def __init__(self, number, suit):
        self.number = number
        self.suit = suit

    def __repr__(self):
        return f"{names[self.number]} of {suits[self.suit]}"

deck = []
for x in range(2,15):
    for i in range(6):
        deck.append(Card(x,i))

def opts(n,d=None):
    functional_deck = deck.copy()
    if d != None:
        for x in d:
            functional_deck.remove(deck[x])
    cards = []
    for x in range(n):
        cd = r.choice(functional_deck)
        cards.append(cd)
        functional_deck.remove(cd)
    return cards

def BakersDozen(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 6:
            return True
    else:
        return False

def RoyalFlush(hand):
    high_cards = [card for card in hand if card.number >= 9]
    sorted_cards = sorted(high_cards, key=lambda card: card.suit)
    grouped_cards = {key: list(group) for key, group in groupby(sorted_cards, key=lambda card: card.suit)}
    potent = list(grouped_cards.values())
    if potent == []:
        return False
    if max(len(x) for x in potent) >= 6:
        return True
    else:
        return False

def Yahtzee(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 5:
            return True
    else:
        return False

def FOAK(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 4:
            return True
    else:
        return False

def Pair(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 2:
            return True
    else:
        return False

def TOAK(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 3:
            return True
    else:
        return False

def TwoPair(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 2:
            if i == 15:
                return False
            for x in range(i+1,15):
                if nums.count(x) >= 2:
                    return True
    else:
        return False

def ThreePair(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 2:
            if i == 15:
                return False
            for x in range(i+1,15):
                if nums.count(x) >= 2:
                    if x == 15:
                        return False
                    for j in range(x+1,15):
                        if nums.count(j) >= 2:
                            return True
    else:
        return False

def RoyalFamily(hand):
    nums = [x.number for x in hand if x.number > 10]
    if len(nums) >= 6:
        return True
    else:
        return False

def AlmostFull(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 3:
            if i == 15:
                return False
            for x in range(i+1,15):
                if nums.count(x) >= 2:
                    return True
    else:
        return False

def FullHouse(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 4:
            if i == 15:
                return False
            for x in range(i+1,15):
                if nums.count(x) >= 2:
                    return True
    else:
        return False

def DoubleTrio(hand):
    nums = [x.number for x in hand]
    for i in range(2,15):
        if nums.count(i) >= 3:
            if i == 15:
                return False
            for x in range(i+1,15):
                if nums.count(x) >= 3:
                    return True
    else:
        return False

def Straight(hand):
    nums = [x.number for x in hand]
    for_each = []
    for x in range(2, 15):
        for_each.append(nums.count(x))
    seqs = [for_each[x:x+6] for x in range(8)]
    for x in seqs:
        if all(i > 0 for i in x):
            return True
    if 2 in nums and 3 in nums and 4 in nums and 5 in nums and 6 in nums and 14 in nums:
        return True
    return False

def Flush(hand):
    suits = [x.suit for x in hand]
    for i in range(2,15):
        if suits.count(i) >= 6:
            return True
    else:
        return False

def StraightFlush(hand):
    sorted_hand = sorted(hand, key=lambda card: card.suit)
    grouped_by_suit = [list(group) for _, group in groupby(sorted_hand, key=lambda card: card.suit)]
    for x in grouped_by_suit:
        if Straight(x):
            return True
    return False

k = 0
for x in range(1_000_000):
    h = opts(9)
    if Straight(h) == True:
        k += 1

def HighCard(hand):
    return True
print(k)
```

```python
hand_rank1 = [None,Pair,TwoPair,TOAK,ThreePair,AlmostFull,Straight,RoyalFamily,FOAK,
             DoubleTrio,FullHouse,Flush,Yahtzee,StraightFlush,BakersDozen,RoyalFlush]
hand_rank = [Pair,TwoPair,TOAK,ThreePair,AlmostFull,Straight,RoyalFamily,FOAK,
             DoubleTrio,FullHouse,Flush,Yahtzee,StraightFlush,BakersDozen,RoyalFlush]

def BestHand(hand):
    #print(hand)
    possible_hands = list(combinations(hand, 6))
    res = [possible_hands] + [[] for x in hand_rank]
    n = 0
    for x in possible_hands:
        for i in hand_rank:
            n += 1
            if i(x) == True:
                res[hand_rank.index(i)+1].append(x)
    res.reverse()
    for x in res:
        if x != []:
            res.reverse()
            a = (hand_rank1[res.index(x)], x)
            if a == None:
                return (None, ChooseTiebreak(None,possible_hands))
            else:
                return (a[0],ChooseTiebreak(a[0],a[1]))




```

```python
def Extract(func,hand):

    if func == Pair:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(2) + 2)
        nums.remove(vals[0])
        nums.remove(vals[0])
        for x in range(4):
            vals.append(max(nums))
            nums.remove(vals[x+1])
        return vals

    if func == TwoPair:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(2) + 2)
        for_each.remove(2)
        vals.append(for_each.index(2) + 3)
        nums.remove(vals[0])
        nums.remove(vals[0])
        nums.remove(vals[1])
        nums.remove(vals[1])
        vals.reverse()
        for x in range(2):
            vals.append(max(nums))
            nums.remove(vals[x+2])
        return vals

    if func == TOAK:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(3) + 2)
        nums.remove(vals[0])
        nums.remove(vals[0])
        nums.remove(vals[0])
        for x in range(3):
            vals.append(max(nums))
            nums.remove(vals[x+1])
        return vals


    if func == ThreePair:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(2) + 2)
        for_each.remove(2)
        vals.append(for_each.index(2) + 1)
        for_each.remove(2)
        vals.append(for_each.index(2))
        vals.reverse()
        return vals

    if func == AlmostFull:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(3) + 2)
        vals.append(for_each.index(2) + 2)
        nums.remove(vals[0])
        nums.remove(vals[0])
        nums.remove(vals[0])
        nums.remove(vals[1])
        nums.remove(vals[1])
        vals.append(max(nums))
        return vals

    if func == FOAK:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(4) + 2)
        for x in range(4):
            nums.remove(vals[0])
        for x in range(2):
            vals.append(max(nums))
            nums.remove(vals[x+1])
        return(vals)


    if func == FullHouse:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(4) + 2)
        vals.append(for_each.index(2) + 2)
        return vals

    if func == DoubleTrio:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(3) + 2)
        for_each.remove(3)
        vals.append(for_each.index(3) + 1)
        vals.reverse()
        return vals

    if func == Yahtzee:
        vals = []
        nums = [x.number for x in hand]
        for_each = []
        for x in range(2, 15):
            for_each.append(nums.count(x))
        vals.append(for_each.index(5) + 2)
        for x in range(5):
            nums.remove(vals[0])
        vals.append(max(nums))
        return vals

    if func == BakersDozen:
        vals = []
        nums = [x.number for x in hand]
        vals.append(max(nums))
        return vals

    if func in [Straight,RoyalFamily,Flush,StraightFlush,None]:
        vals = []
        nums = [x.number for x in hand]
        for x in range(6):
            vals.append(max(nums))
            nums.remove(vals[x])
        return vals

    if func == RoyalFlush:
        return [1]

def compare_lists(lists):
    contenders = lists
    k = 0
    while len(contenders)>1:
        if k == len(contenders[0]):
            return None
        ma = max([i[k] for i in contenders])
        contenders = list(filter(lambda x: x[k]==ma,contenders))
        k += 1
    return contenders[0]

def ChooseTiebreak(func, hands):
    scores = [Extract(func,hand) for hand in hands]
    top = compare_lists(scores)
    if top == None:
        return r.choice(hands)
    else:
        return hands[scores.index(top)]




```

# Starting Hand Simulations

```python
def Draw(n):
    functional_deck = deck.copy()
    hands  = []
    communal = []
    for x in range(n):
        h = []
        for i in range(3):
            cd = r.choice(functional_deck)
            h.append(cd)
            functional_deck.remove(cd)
        hands.append(h)
    for x in range(6):
        cd = r.choice(functional_deck)
        communal.append(cd)
        functional_deck.remove(cd)
    return [x + communal for x in hands]

def Winner(p1,p2):
    r1 = BestHand(p1)
    r2 = BestHand(p2)
    if hand_rank1.index(r1[0]) > hand_rank1.index(r2[0]):
        return 1
    elif hand_rank1.index(r1[0]) < hand_rank1.index(r2[0]):
        return 0
    else:
        ak = ChooseTiebreak(r1[0],[r1[1],r2[1]])
        if ak == r1[1]:
            return 1
        else:
            return 0


```

```python
import pickle

def Classify(hand):
    x = [i.suit for i in hand]
    y = [i.number for i in hand]
    if x[0] == x[1] and x[1] == x[2]:
        return (y,[1,1,1])
    elif x[0] == x[1]:
        return (y,[1,1,0])
    elif x[0] == x[2]:
        return (y,[1,0,1])
    elif x[1] == [2]:
        return (y,[0,1,1])
    else:
        return (y,[0,0,0])

file_path = "/content/drive/MyDrive/poker_sim.pkl"

starting_hands = list(combinations(deck,3))
se = [Classify(x) for x in starting_hands]
sed = []
[sed.append(x) for x in se if x not in sed]
# results = [(x,[]) for x in sed]
with open(file_path, "rb") as file:
    results = pickle.load(file)






for x in range(200_000):
    if x % 1000 == 0:
        print(x)
    d = Draw(2)
    sh1 = Classify(sorted(d[0][:3],key=lambda x: x.number))
    sh2 = Classify(sorted(d[1][:3],key=lambda x: x.number))
    if Winner(d[0],d[1]) == 1:
        results[sed.index(sh1)][1].append(1)
        results[sed.index(sh2)][1].append(0)
    else:
        results[sed.index(sh1)][1].append(0)
        results[sed.index(sh2)][1].append(1)
```
